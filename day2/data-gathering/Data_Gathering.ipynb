```python
import paho.mqtt.client as mqtt
import pandas as pd

client=mqtt.Client()
client.connect('broker.hivemq.com',1883)
print('Broker Connected')

client.subscribe('rcew/ml')
data=[]
i=0
def notification(client,userdata,msg):
  global data,i
  t=msg.payload.decode('utf-8')
  t=t.split(' ')
  hum=int(t[1][:-1])
  temp=int(t[-1][:-1])
  label=0
  if((hum>40 and hum<=60)):
    label=1
  elif((hum>60 and hum<=80)):
    label=2
  else:
    label=0
  #print(hum)
  #print(temp)
  dummy=[]
  dummy.append(hum)
  dummy.append(temp)
  dummy.append(label)
  data.append(dummy)
  print(i)
  i+=1
  if(i==50):
    df=pd.DataFrame(data)
    i=0
    df.to_csv('data.csv')


client.on_message=notification
client.loop_forever()
```
